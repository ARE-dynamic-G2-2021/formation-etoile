```python
from tkinter import *
import math
import random
import numpy as np



# les variables suivantes seront utilisées de manière globale :
n = 100
flag =0                  # commutateur
t = 0                # temps
x0,y0 = 500,500     # centre du cercle
#r = 250         # rayon
LARGEUR = 1000
HAUTEUR = 1000

# Création du widget principal:
fen1 = Tk()
fen1.title("Exercice d'animation avec Tkinter")


# position initiale des balles
X = []
Y = []
for i in range(n):
    X.append(LARGEUR*np.random.rand())
    Y.append(HAUTEUR*np.random.rand())

# création des widgets :

can1 = Canvas(fen1,bg='dark grey',height=HAUTEUR, width=LARGEUR)
can1.pack(side=LEFT, padx =5, pady =5)

#création des objets graphiques:
Balle=[]
for i in range(n):
    Balle.append(can1.create_oval(X[i]-10,Y[i]-10,X[i]+10,Y[i]+10,width=1,fill='magenta'))

can1.create_oval(x0-10,y0-10,x0+10,y0+10,width=1,fill='blue')
def move():
    "déplacement de la balle"
    global t
    delta = 0.2
    for i in range(n): 
        r = math.sqrt((x0-X[i])*(x0-X[i]) + (y0-Y[i])*(y0-Y[i]))
        X[i] = x0 + r * math.cos(t) 
        Y[i] = y0 + r * math.sin(t)
        t += delta
        if t > 2*math.pi:
            t = 0
            
        can1.coords(Balle[i],X[i]-10,Y[i]-10,X[i]+10,Y[i]+10)
    if flag >0:
        fen1.after(100,move)

def stop_it():
    "arret de l'animation"
    global flag   
    flag =0
 
def start_it():
    "démarrage de l'animation"
    global flag
 
    if flag ==0:       # pour ne lancer qu'une seule boucle
     flag =1
     move()

#créations des widgets boutons:
bou1 = Button(fen1,text='Quitter', width =8, command=fen1.destroy)
bou1.pack(side=BOTTOM)
bou2 = Button(fen1, text='Démarrer', width =8, command=start_it)
bou2.pack()
bou3 = Button(fen1, text='Arrêter', width =8, command=stop_it)
bou3.pack()
 
# démarrage du réceptionnaire d'évènements (boucle principale) :
fen1.mainloop()
```

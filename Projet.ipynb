```python
# script animation.py
#(C) Fabrice Sincère
from tkinter import *
import random
import math
import numpy as np

Rayon = 120
def Cercle():
    """ Dessine un cercle de centre (x,y) et de rayon r """
    global Arret 
    t = random.uniform(0,2*math.pi)
    R = random.uniform(0,Rayon)
    x = R*math.cos(t) + Largeur/2
    y = R*math.sin(t) + Hauteur/2
    r = 2
    Canevas.create_oval(x-r, y-r, x+r, y+r, outline='white', fill='white')
    if Arret == False:
        # appel de la fonction Cercle() après une pause de 50 millisecondes
        Mafenetre.after(2,Cercle) 

def pied() : 
    global Arret
    L=[]
    for i in range(0,9) : 
        L.append(i*(math.pi/4))
        
    #print(L)
    d = random.randint(0,7)
    x0 = Rayon*math.cos(L[d]) + Largeur/2
    y0 = Rayon*math.sin(L[d]) + Hauteur/2
    t = random.uniform(0,math.pi/8)
    R = random.uniform(0,Rayon) 
    if (L[d] == math.pi/2) : 
        x = -R*math.cos(t) + x0
        y = R*math.sin(t) + y0 
    elif (L[d] == 3*math.pi/2) : 
        x = R*math.cos(t) + x0
        y = -R*math.sin(t) + y0 
    elif (L[d] > math.pi/2 and L[d] < 3*math.pi/2) : 
        x = -R*math.cos(t) + x0
        y = -R*math.sin(t) + y0
    else : 
        x = R*math.cos(t) + x0
        y = R*math.sin(t) + y0
    r = 2
    Canevas.create_oval(x-r, y-r, x+r, y+r, outline='white', fill='white')
    if Arret == False :
        Mafenetre.after(2,pied)

def Arreter():
	""" Arrêt de l'animation """
	global Arret
	Arret = True

def Demarrer():
    """ Démarre l'animation """
    global Arret
    Canevas.delete(ALL)
    if Arret == True:
        Arret = False
        Cercle() # un seul appel à cette fonction
        pied()
Arret = True

# Création de la fenêtre principale (main window)
Mafenetre = Tk()
Mafenetre.title('Animation')

# Création d'un widget Canvas
Largeur = 500
Hauteur = 500
Canevas = Canvas(Mafenetre, width = Largeur, height =Hauteur, bg ='black')
Canevas.pack(padx =5, pady =5)

# Création d'un widget Button (bouton Démarrer)
BoutonGo = Button(Mafenetre, text ='Démarrer', command = Demarrer)
BoutonGo.pack(side = LEFT, padx = 10, pady = 10)

# Création d'un widget Button (bouton Arrêter)
BoutonArreter = Button(Mafenetre, text ='Arrêter', command = Arreter)
BoutonArreter.pack(side = LEFT, padx = 5, pady = 5)

# Création d'un widget Button (bouton Quitter)
BoutonQuitter = Button(Mafenetre, text ='Quitter', command = Mafenetre.destroy)
BoutonQuitter.pack(side = LEFT, padx = 5, pady = 5)

Mafenetre.mainloop()
```
